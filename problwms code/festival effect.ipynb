```python
# We look at a 3-day window around the festival (Day Before, Festival, Day After)
def get_day_type(date_obj):
    date_str = date_obj.strftime('%Y-%m-%d')
    
    # Check if this date is a festival
    if date_str in festival_calendar:
        return 'Festival Day'
    
    # Check if it's near a festival (Window of +/- 2 days)
    for fest_date in festival_calendar:
        fest_dt = pd.to_datetime(fest_date)
        diff = (date_obj - fest_dt).days
        if abs(diff) <= 2:
            return 'Festival Window'
            
    return 'Normal Day'

df['Day_Type'] = df['date'].apply(get_day_type)
```

```python
summary = df.groupby('Day_Type')['Total_Updates'].mean().reset_index()

# Calculate Lift/Drop
normal_avg = summary[summary['Day_Type'] == 'Normal Day']['Total_Updates'].values[0]
summary['Impact_%'] = ((summary['Total_Updates'] - normal_avg) / normal_avg) * 100

print("--- Festival Impact Report ---")
print(summary)
```

```python
# 4. Visualization
plt.figure(figsize=(10, 6))
colors = {'Normal Day': 'grey', 'Festival Window': 'orange', 'Festival Day': 'red'}

sns.barplot(x='Day_Type', y='Total_Updates', data=summary, palette=colors)
plt.axhline(normal_avg, color='black', linestyle='--', label='Normal Daily Average')

plt.title('Do Festivals Kill Productivity or Create Demand?', fontsize=14)
plt.ylabel('Average Daily Updates')
plt.legend()
plt.show()
```
